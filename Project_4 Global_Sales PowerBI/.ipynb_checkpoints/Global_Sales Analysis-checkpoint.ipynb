# Looking for Seasonal Trends  and sale patterns 

```python
#Trying to identify seasonal patterns over the year 

monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()

# Filter for your specific date range
start_date = '2015-01-01'
end_date = '2018-12-31'
monthly_sales = monthly_sales[(monthly_sales.index >= start_date) & (monthly_sales.index <= end_date)]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values)
plt.title('Monthly Sales from 2015 to 2018')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
#Iderying to identify patterns for the sales over a month

df['Day of Month'] = df['Order Date'].dt.day

# Group by the day of the month and calculating the mean sales 
average_daily_sales = df.groupby('Day of Month')['Sales'].mean()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(average_daily_sales.index, average_daily_sales.values)
plt.title('Average Sales by Day of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Average Sales')
plt.xticks(range(1, 32))  # Set x-ticks for each day
plt.grid(True)
plt.show()
```

```python
#Trying to identify patterns for the sales over a week
# Create a new column for the day of the week
df['Day of the Week'] = df['Order Date'].dt.day_name()

# Group by the day of the week and calculate the mean sales
average_weekday_sales = df.groupby('Day of the Week')['Sales'].mean().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plotting - using line plot this time
plt.figure(figsize=(12, 6))
plt.plot(average_weekday_sales.index, average_weekday_sales.values, marker='o')  # Added marker for clarity
plt.title('Average Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Sales')
plt.grid(True)
plt.show()
```

```python
#The above results where suprising so I decided to perform a chcek on the days with most sales: Wednesday and Thursday.
```

```python
# Count of transactions per day of the week
count_weekday_sales = df.groupby('Day of the Week')['Sales'].count().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(count_weekday_sales.index, count_weekday_sales.values, marker='o', color='red')
plt.title('Count of Transactions by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count of Transactions')
plt.grid(True)
plt.show()
```

```python
#Suprisingly the days with least sales are Wednesday and Thursday and they still generate most sales
# **BUT** this would make sense only if the products sold during Wednesday Thursday would be the most expensive ones
#  so I will check which products are most sold during these days to make sure the data is correct 

weekly_product_sales = df.groupby(['Day of the Week', 'Product Name']).size()

# Sort the sales within each day group in descending order and take the top 3
top_weekly_products = weekly_product_sales.groupby('Day of the Week').nlargest(3).reset_index(level=0, drop=True)

# Now 'top_weekly_products' has the top 3 products for each day of the week based on the count of transactions
print(top_weekly_products)
```

```python
# Group by 'Day of the Week' and 'Product Name', then sum the sales for each product
weekly_product_sales = df.groupby(['Day of the Week', 'Product Name'])['Sales'].sum()

# Sort the sales within each day group in descending order and take the top 3
top_weekly_products = weekly_product_sales.groupby('Day of the Week').nlargest(3).reset_index(level=0, drop=True)

# Print the top 3 products with their sales for each day of the week
print(top_weekly_products)
```

### Sales data  distriution by State

```python
plt.figure(figsize=(14, 6))

sns.boxplot(x='State', y='Sales', data=df)
plt.xticks(rotation=90)  # Rotate the state labels for better readability
plt.title('Sales Distribution by State')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```

### Sales By State

```python
# Group by 'State' to calculate the sum of sales
sales_per_state = df.groupby('State').sum(numeric_only=True).sort_values("Sales", ascending=False)


# Assuming 'sales_per_state' is your DataFrame with 'Sales' being the summed total sales per state
sales_per_state['Sales'].plot(kind='bar', figsize=(20, 10))
plt.title('Total Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)  # Rotate the state labels for better readability
plt.show()
```

### Most sold and least sold products 


```python
#Finding Most Sold and Least Sold products

# Calculate the sum of sales for each product
most_sold_products = df.groupby("Product Name").sum(numeric_only=True).sort_values("Sales", ascending=False)

# Count the number of times each product has been sold 
# Converting the Series to a DataFrame
times_sold = df['Product Name'].value_counts()
times_sold_df = times_sold.rename_axis('Product Name').reset_index(name='Times Sold')

# Merging the new 'Times Sold column' with the the most_sold_product df
most_sold_products = most_sold_products.merge(times_sold_df, on='Product Name', how='left')


most_sold_products.head(10)


```

```python
#Least sold products, accesing most sold products and reversing so they are sorted by the least sold 
least_sold_product = most_sold_products[-10:][::-1]
least_sold_product
```
